```python
import pandas as pd
import seaborn as sb
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from IPython.display import display
%matplotlib inline
```

```python
now=dt.datetime.now()
print(now)
```

```python
t=np.linspace(0,2*3600, 2*3600*10+1)
print(t)
```

```python
dt_t=[]
for time in t:
    dt_t.append(now+dt.timedelta(seconds=time))
print(dt_t[0:10])
```

```python
time_index=pd.date_range(now,periods=2*36000, freq='0.1S')
print(time_index)
```

```python
#pandas - timeseries , dataframe
```

```python
data_1=pd.Series(np.random.randn(len(time_index)), index=time_index)
print(data_1)
```

```python
data_1.name='Data 1'
data_1.plot()
plt.legend()
```

```python
data_1[0:100].plot()
data_1.rolling('1S').mean().plot()
data_1.rolling('0.5S').mean().plot()
data_1.rolling('2S').mean().plot()
plt.legend()
```

```python
data_1[0:100].plot()
data_1[0:100].resample('1S').mean().plot()
data_1[0:100].resample('0.2S').mean().plot()
```

```python
data_1[0:100].plot()
data_1[0:100].resample('1S').sum().plot()
```

```python
##dataframe
```

```python
data_2=pd.Series(np.sin(t[:-1])+np.random.randn(len(t[:-1]))*0.1, index=time_index)
data_3=pd.Series(np.sin(t[:-1]+1)+np.random.randn(len(t[:-1]))*0.1, index=time_index)
data_4=pd.Series(np.sin(t[:-1])+np.random.randn(len(t[:-1]))*0.5, index=time_index)
```

```python
df1=data_1.to_frame()
df1['Data 2']=data_2
df1['Data 3']=data_3
df1['Data 4']=data_4

display(df1)
```

```python
df1[0:100].plot()
plt.legend
```

```python
df1[0:100].resample('0.5S').mean().plot()
```

```python
df1[['Data 1', 'Data 4']].resample('0.2S').mean()[0:50].plot() # seperate different lines
```

```python
df1[df1['Data 1'] >df1['Data 2']][0:100].plot()
```

```python
now=dt.datetime.now()
time_index=pd.date_range(now,periods=2*36000, freq='0.1S')
t=np.linspace(0,2*3600, 2*3600*10+1)
data_1=pd.Series(np.random.randn(len(time_index)), index=time_index)

data_2=pd.Series(np.sin(t[:-1])+np.random.randn(len(t[:-1]))*0.1, index=time_index)
data_3=pd.Series(np.sin(t[:-1]+1)+np.random.randn(len(t[:-1]))*0.1, index=time_index)
data_4=pd.Series(np.sin(t[:-1])+np.random.randn(len(t[:-1]))*0.5, index=time_index)
df1=data_1.to_frame()
df1['Data 2']=data_2
df1['Data 3']=data_3
df1['Data 4']=data_4
sb.pairplot(df1)
```

```python
sb.pairplot(df1.resample('10S').mean())
```

```python
g=sb.PairGrid(df1.resample('60S').mean())
g.map_diag(sb.kdeplot)
g.map_offdiag(sb.kdeplot,cmap='Blues_d',n_level=6)
```

```python
sb.distplot(df1['Data 1'])
```

```python
sb.kdeplot(df1['Data 2'],shade=True, color='g')
```

```python
sb.jointplot(x='Data 3', y='Data 2', data=df1)
```

```python
sb.jointplot(x='Data 3', y='Data 2', data=df1, kind='hex', color='r')
```

```python
sb.jointplot(x='Data 3', y='Data 2', data=df1.resample('60S').mean(), kind='hex', color='r')
```
