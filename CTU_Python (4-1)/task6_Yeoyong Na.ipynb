### a) Simple Moving Average filter
Create a function "avgfilter" witch accept two parameters - y = data vector, n = number of samples for the filter. The function avgfilter returns a numpy array witch contains data filtered by simple moving average filter - each new data point is calculated as unweightened mean of "n" previous input data points.

```python
# Your solution here
# Plot original data and filtered data in one matplotlib figure

def avgfilter(y,n):
    import numpy as np
    new=np.array([])
    for i in range(0,len(y)-n+1):
        new = np.append(new, np.mean(y[i:i + n]))
    for j in range(0,len(y)-len(new)):    
        new = np.insert(new,0,0,axis=None)        
    return new

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
t = np.linspace(0,100,1000)
data = np.sin(2.25*t)+np.sin(0.5*t)+np.cos(3*t)+np.random.randn(len(t))
plt.plot(t, data)
plt.plot(t, avgfilter(data,3))
plt.xlabel('time [s]')
plt.ylabel('data')
plt.grid(True)        
```

### a) Data Resampling
Create a function "resample" witch accept two parameters - t = time vector, y = data vector, s = new_sampling period. The function resample returns two numpy arrays witch contains new resampled time vector and new resampled data vector. Use the linear approximation for data resampling.

```python
# Your solution here
# Plot original data and resampled data in one matplotlib figure

def resample(t,y,s):
    new_t=t*s
    new_y=y*s
    return (new_t,new_y)

import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
t = np.linspace(0,100,1000)
data = np.sin(2.25*t)+np.sin(0.5*t)+np.cos(3*t)+np.random.randn(len(t))
plt.plot(t, data)
(new_t,new_y)=resample(t,data,0.2)
plt.plot(new_t,new_y)
plt.xlabel('time [s]')
plt.ylabel('data')
plt.grid(True)
```
